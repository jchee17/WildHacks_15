```python
import time
import pandas as pd
import statsmodels.api as sm
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
from patsy import dmatrices
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score

# read the data in
df = pd.read_csv("/home/jerry/GitHub/WildHacks_15/113926557_T_ONTIME.csv")
```

```python
df.describe()
#df.std()
```

```python
# select variables
data = df.drop('CANCELLED', axis = 1)
data = data.drop('DIVERTED', axis = 1)
data = data.drop('CRS_ELAPSED_TIME', axis = 1)
data = data.drop('AIR_TIME', axis=1) 
data = data.drop('CARRIER_DELAY', axis = 1)
data = data.drop('WEATHER_DELAY', axis = 1)
data = data.drop('NAS_DELAY', axis = 1)
data = data.drop('SECURITY_DELAY', axis = 1)
data = data.drop('LATE_AIRCRAFT_DELAY', axis = 1)
data = data.drop('Unnamed: 32', axis = 1)
data = data.drop('ACTUAL_ELAPSED_TIME', axis = 1)
```

```python
data.describe()
```

```python
# create dataframes with an intercept column and dummy variables 
y, X = dmatrices('ARR_DELAY ~ C(YEAR) + C(QUARTER) + \
        C(MONTH) + C(DAY_OF_MONTH) + C(DAY_OF_WEEK) + \
        C(AIRLINE_ID) + CRS_DEP_TIME + DEP_TIME + DEP_DELAY + \
        CRS_ARR_TIME + FLIGHTS + DISTANCE',
                 data, return_type='dataframe')
print X.columns
```

```python
# flatten y into a 1-D array
y = np.ravel(y)
```

```python
# Linear Regression Model, fit X and y
start_time = time.time()

model = LinearRegression()
model = model.fit(X,y)

elapsed_time = time.time() - start_time
print elapsed_time

# check accuracy on the training set
model.score(X,y)
```

```python
pd.DataFrame(zip(X.columns, np.transpose(model.coef_)))
```

```python
# evaluate the model by splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model2 = LinearRegression()
model2.fit(X_train, y_train)
```

```python
# predict for test set
predicted = model2.predict(X_test)
print predicted
```

```python
# evaluate the model using 10-fold cross-validation
scores = cross_val_score(LinearRegression(), X, y, scoring='accuracy', cv=10)
print scores
print scores.mean()
```
